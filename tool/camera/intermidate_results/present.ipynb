```python
from matplotlib import pyplot as plt
import numpy as np

files = [
    'color_raw.png',
    'depth_raw.png',
    'depth_post.png',
    'mask.png',
    'depth_ground.png',
    'depth_adjust.png',
]

names = [
    'RGB',
    'Depth Raw',
    'Depth Post',
    'Mask',
    'Depth Ground',
    'Depth Adjust',
]

## plot the above file in 2x3 grid and save it as 'output.png', give each plot a title
fig, axs = plt.subplots(2, 3, figsize=(18, 10))
for i, file in enumerate(files):
    img = plt.imread(file)
    axs[i//3, i%3].imshow(img)
    axs[i//3, i%3].set_title(names[i], fontsize=20)
    axs[i//3, i%3].axis('off')

plt.tight_layout()
plt.savefig('output.png')
```
